```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

### Scraping the Table
First we import the HTML using the requests library, and then pass it through as a BeautifulSoup object.
From there, I use soup.find() to search for the tables on the page, and choose the item from the list to read as a dataframe.

### Cleaning up the Table
Then, I'll drop the Borough's labeled as "Not assigned" and reset the index to fix the row numbering. Then, I reassign the values of the "Not assigned" neighborhoods to be the name of the Borough.

```python
url=requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M").text

soup=BeautifulSoup(url, features="lxml")
# print(soup)

test=soup.find('table', {"class":"wikitable"})
# print(test)

df=pd.read_html(str(test))
df=df[0]
df=df.loc[df['Borough'] != "Not assigned"]
df = df.reset_index(drop=True)
df.Neighbourhood[df.Neighbourhood=='Not assigned'] = df.Borough
df
```

### Combine the Neighborhoods
Now, using the groupby method, I will combine the neighborhood cells based on matching postcode/boroughs, and again, reset the index to fix the row numbering.

```python
df = df.groupby(['Postcode','Borough'])['Neighbourhood'].apply(', '.join).reset_index()
df
```

### Shape Method
USing the df.shape method, we see there are 103 rows and 3 columns.

```python
df.shape
```

### Imporitng the CSV
Since the module wasn't working, I'm importing the provided CSV file, naming it "geo".

```python
geo = pd.read_csv("https://cocl.us/Geospatial_data")
```

### Adding the Latitude and Longitude
I'll now perform a join with the geo dataframe's index set to be the postal code, on the original dataframe's Postcode column. 

```python
df = df.join(geo.set_index('Postal Code'), on='Postcode')
df
```

### Now let's map the neighborhoods.

```python
# create map of Toronto using latitude and longitude values
map_TO = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, borough, neighborhood in zip(df['Latitude'], df['Longitude'], df['Borough'], df['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_TO)  
    
map_TO
```
